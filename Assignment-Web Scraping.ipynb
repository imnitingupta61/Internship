# Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.

```python
page = requests.get("https://coreyms.com/")
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
head = []

for i in soup.find_all("h2"):
    head.append(i.text)
head
```

```python
tm =[]

for i in soup.find_all("time", class_="entry-time", itemprop="datePublished"):
    tm.append(i.text)
    
tm
```

```python
con=[]

for i in soup.find_all('div', class_="entry-content", itemprop="text"):
    con.append(i.text.replace("\n", ""))
    
con
```
