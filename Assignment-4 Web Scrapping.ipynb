# Ques-1 Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
driver = webdriver.Chrome(r"C:\Users\NITIN GUPTA\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
rank= []
name=[]
artist=[]
date=[]
views=[]

ra= driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
for r in ra:
    rank.append(r.text)
na= driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
for n in na:
    name.append(n.text)
ar= driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
for a in ar:
    artist.append(a.text)
    
dt= driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
for d in dt:
    date.append(d.text)
    
vw= driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
for v in vw:
    views.append(v.text)
```

```python
df= pd.DataFrame({'Rank':rank, "Name":name, "Artist":artist, "Date":date, "Views":views})
df.head(30)
```

```python
df = pd.DataFrame({'Rank':rank, 'State':state, "GSDP(18-19)":gsdp_18_19, "GSDP(19-20)":gsdp_19_20, "Share":share, "GDP":gdp})
df
```
